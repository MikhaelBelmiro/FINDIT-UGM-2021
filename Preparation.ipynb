```python
import pandas as pd
pd.set_option('display.max_columns', 50)

import seaborn as sn
import matplotlib as mpl
from matplotlib import pyplot as plt
```

```python
train = pd.read_csv('public-train.csv',delimiter='|')
train.sample(5)
```

```python
train.info()
```

```python
for i, column in enumerate(train.columns):
    if i == 1:
        continue
    print(f'Unique values dari {column} adalah:')
    print(train[column].unique())
    print()
```

```python
fig, size = plt.subplots(1,1, figsize=(16,16))
sn.heatmap(train.corr(),annot=True)
```

```python
fig, size = plt.subplots(1,1, figsize=(16,16))
sn.heatmap(train.corr('spearman'),annot=True)
```
